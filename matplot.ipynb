```python
import matplotlib.pyplot as plt
import numpy as np
x=([3,4,5,6,7,8])
y=([30,40,50,60,70,80])
plt.subplot(2,1,1)
plt.plot(x,marker='o',ms='10',c='g')
plt.plot(y,linestyle='dotted',c='y')
plt.title("Rollno's and their eamcet ranks",c='r')
plt.xlabel("rollno",c='g',)
plt.ylabel("eamcetranks",c='y')
plt.show()
```
